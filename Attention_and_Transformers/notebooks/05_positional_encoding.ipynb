## 1. The Problem: Attention is Position-Agnostic

Self-attention doesn't inherently know about token order:
- Attention is a sum of values weighted by similarity
- No mechanism to encode position information
- Same input in different order = same output

### Solution: Add Position Information to Embeddings

Instead of just word embeddings, add position information:

```
Final Embedding = Word Embedding + Positional Encoding
```

## 2. Sinusoidal Positional Encoding

### Formula

$$PE_{(pos, 2i)} = sin(pos / 10000^{2i/d_{model}})$$
$$PE_{(pos, 2i+1)} = cos(pos / 10000^{2i/d_{model}})$$

Where:
- `pos` = position in sequence (0, 1, 2, ...)
- `i` = dimension index (0, 1, 2, ...)
- `d_model` = model dimension

````python
def sinusoidal_positional_encoding(seq_len, d_model):
    """
    Generate sinusoidal positional encoding.
    
    Args:
        seq_len: Maximum sequence length
        d_model: Model dimension
    
    Returns:
        pe: (seq_len, d_model) positional encoding
    """
    pe = np.zeros((seq_len, d_model))
    position = np.arange(0, seq_len).reshape(-1, 1)
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    
    pe[:, 0::2] = np.sin(position * div_term)
    pe[:, 1::2] = np.cos(position * div_term)
    
    return pe

# Example
seq_len = 100
d_model = 512

pe = sinusoidal_positional_encoding(seq_len, d_model)

print(f"Positional Encoding shape: {pe.shape}")
print(f"First position: {pe[0, :10]}")
print(f"Second position: {pe[1, :10]}")
print(f"Tenth position: {pe[10, :10]}")

print(f"\nProperties:")
print(f"  Min: {pe.min():.4f}, Max: {pe.max():.4f}")
print(f"  Mean: {pe.mean():.4f}, Std: {pe.std():.4f}")
````
