## 1. Major Transformer Applications

### 1. Machine Translation
- **Architecture**: Encoder-Decoder (Transformer-base)
- **Input**: Source language tokens
- **Output**: Target language tokens
- **Examples**: Google Translate, DeepL

### 2. Text Classification
- **Architecture**: Encoder only (BERT-style)
- **Input**: Document tokens
- **Output**: Class probabilities
- **Examples**: Sentiment analysis, topic classification

### 3. Language Modeling
- **Architecture**: Decoder only (GPT-style)
- **Input**: Token sequence
- **Output**: Next token prediction
- **Examples**: GPT-3, GPT-4

### 4. Named Entity Recognition (NER)
- **Architecture**: Encoder + CRF layer
- **Input**: Document tokens
- **Output**: Entity tags per token
- **Examples**: Extract names, locations, organizations

### 5. Question Answering
- **Architecture**: Encoder-based (SQuAD task)
- **Input**: Question + context passage
- **Output**: Answer span positions
- **Examples**: SQuAD, MS MARCO

### 6. Summarization
- **Architecture**: Encoder-Decoder
- **Input**: Long document
- **Output**: Short summary
- **Examples**: T5, BART

## 4. Transfer Learning

Use pretrained models for downstream tasks

```python
print("Transfer Learning Strategy:")
print()
print("1. Pretrain on large corpus (expensive)")
print("   - Learn general language patterns")
print("   - Understand grammar, semantics")
print()
print("2. Fine-tune on downstream task (cheap)")
print("   - Use small labeled dataset")
print("   - Adapt to specific task")
print()
print("Common pretrained models:")
print("  - BERT: Bidirectional encoder (classification)")
print("  - GPT: Autoregressive decoder (generation)")
print("  - T5: Encoder-Decoder (sequence-to-sequence)")
print("  - RoBERTa: Improved BERT")
print("  - ELECTRA: More efficient pretraining")
```

## 5. When to Use Transformers

### ✅ Use Transformers When:
- You have text data with long-range dependencies
- You have enough compute and data
- You need state-of-the-art performance
- Transfer learning can help (pretrained models available)
- You need interpretability (attention visualization)

### ❌ Don't Use Transformers When:
- Very small dataset (< 1000 samples)
- Very short sequences (< 10 tokens)
- Limited compute (CPU only)
- Real-time inference required (too slow)
- Simple task (logistic regression sufficient)

### 🚀 Transformer Alternatives
- **RNN/LSTM**: Simpler, faster, less memory
- **CNN**: Good for local patterns
- **Hybrid**: Combine with other architectures
